#**Qu es GeoPandas?**

**GeoPandas** es un proyecto open source cuyo objetivo es facilitar el trabajo con **datos geoespaciales**

*GeoPandas dota a Python de acceso a una serie de funcionalidades realmente 
tiles para realizar anlisis geoespacial, gestionar geoinformacin y crear salidas cartogrficas de nuestros datos geogrficos.* 


*Combina las capacidades de pandas y shapely, proporcionando operaciones geoespaciales en pandas y una interfaz de alto nivel para mltiples geometras a shapely. GeoPandas le permite realizar fcilmente operaciones en python que de otro modo requeriran una base de datos espacial como PostGIS.*

* Se basa en la librera **Pandas** de Python, adaptando varias de sus clases para dar soporte a datos con geometras asociadas

* Adems de Pandas, tambin se sustenta sobre la librera **Shapely** que permite llevar a cabo operaciones geoespaciales con las geometras asociadas a esos datos

* Geopandas depende adems de **Fiona** para el acceso a archivos y de **Matplotlib** para el plotting

#**GeoPandas y Pandas**

**GeoPandas**, como su nombre indica, ampla la biblioteca de ciencia de datos **Pandas** aadiendo soporte para datos geoespaciales.

GeoPandas implementa dos estructuras de datos principales: `GeoSeries` y  `GeoDataFrame`. Estas son subclases de `pandas.Series` y `pandas.DataFrame`, respectivamente.

## **GeoSeries**

Una `GeoSerie` es esencialmente un vector en el que cada entrada es un conjunto de formas que corresponden a una observacin. 

Una entrada puede consistir en una sola forma (como un nico polgono) o en varias formas que deben considerarse como una observacin (como los diferentes polgonos que componen el Estado de Jalisco o un pas como Mxico)

**GeoPandas** tiene tres clases bsicas de objetos geomtricos (que en realidad son objetos de *shapely*):

* Puntos / Multipuntos   (centroides)

* Lneas / Multi-Lneas

* Polgonos / Multi-Polgonos

Algunos atributos de `GeoSeries`:
* `area`: rea de la forma

* `bounds`: tupla de coordenadas mximas y mnimas en cada eje para cada forma

* `total_bounds`: tupla de coordenadas mximas y mnimas en cada eje para toda la GeoSeries

* `geom_type`: tipo de geometra

* `is_valid`: comprueba si las coordenadas forman una figura geomtrica razonable 

Algunos metodos de `GeoSeries`
* `distance()`: devuelve Series con la distancia mnima de cada entrada a otra

* `centroid`: devuelve GeoSeries de centroides

* `representative_point()`: devuelve GeoSeries de puntos que se garantiza que estn dentro de cada geometra. NO devuelve centroides.

* `to_crs()`: cambia el sistema de referencia de coordenadas

* `plot()`: traza GeoSeries

* `geom_almost_equals()`: te dice si la forma es casi igual a otra

* `contains()`: te dice si la forma est contenida dentro de otra

* `intersects()`: te dice si la forma interseca a otra

## **GeoDataFrame**

dataframe1.jpg

La estructura de datos central en GeoPandas es `geopandas.GeoDataFrame`, que puede almacenar columnas de geometra y realizar operaciones espaciales. 

Un `GeoDataFrame` es una estructura de datos tabular que contiene una `GeoSerie`

El `geopandas.GeoSeries`, como ya se explico, maneja las geometras. Por tanto, el `GeoDataFrame` es una combinacin de `pandas.Series`, con datos tradicionales (numricos, booleanos, texto, etc.), y `geopandas.GeoSeries`, con geometras (puntos, polgonos, etc.). 

Puede tener tantas columnas con geometras como desee, cada `GeoSerie` puede contener cualquier tipo de geometra, sin embargo, slo una `GeoSerie` en un `GeoDataFrame` se considera la geometra activa, lo que significa que todas las operaciones geomtricas aplicadas a un `GeoDataFrame` operan sobre esta columna activa.
